```python

%matplotlib  widget

from astropy.io import fits
import matplotlib.pyplot as plt
from  matplotlib.colors import LogNorm
import numpy as np
import pandas as pd
import scipy as sc
```

```python
file_name='m87_broadband.fits'
mla=fits.open(file_name)
```

```python
mla.info()
```

```python
mdata=mla[0].data
header=mla[0].header

print(header)

```

```python
plt.imshow(mdata,cmap='inferno',origin='lower',norm=LogNorm(vmin=1e-5,vmax=np.max(mdata)))

plt.colorbar(label='intensity')
plt.title('M87')
plt.show()
```

```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np

mdul = fits.open('m87_broadband.fits')
data = mdul[0].data
plt.figure()
plt.hist(data.flatten(), bins=20, range=(np.percentile(data, 10), np.percentile(data, 90)))
plt.xlabel('pixel intensity')
plt.ylabel('frequency')
plt.show()
mdul.close()
```
